```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

# Scatter Plot

```python
plt.scatter(data['Month'], data['Quantity (000 Metric Tonnes)'])
plt.xticks(rotation = 45)
plt.xlabel('Month of the Year')
plt.ylabel('Quantity (000 Metric Tonnes)')
plt.title("Scatter plot for Oil Company quantity suppy per month")
plt.show()
```

# Box plot

```python
quantity = data['Quantity (000 Metric Tonnes)'].dropna().to_numpy()

# Plot
fig = plt.figure(figsize=(10, 7))
plt.boxplot(quantity)
plt.title('Boxplot of Quantity')
plt.show()
```

# Line Plot

```python
# Group by Month and sum the Quantity
monthly_totals = data.groupby('Month')['Quantity (000 Metric Tonnes)'].sum().reset_index()

print(monthly_totals)
```

```python
plt.bar([x for x in range(1,13)], monthly_totals['Quantity (000 Metric Tonnes)'].to_numpy(), color ='maroon', 
        width = 0.4)
 
plt.xlabel("Months")
plt.ylabel("Total Quantity Oil Produced")
plt.title("Line Plot")
plt.show()
```
